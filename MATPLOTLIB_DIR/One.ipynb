# Pyplot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,3,6,34,73,5]
y = [24,3,64,2,66,22]


plt.xlabel('X here')
plt.ylabel('Y here')
plt.title("My data chart")

plt.plot(x,y,color='green', linewidth=4, linestyle='dotted')
```

```python
x = list(range(1,21,2))
y=2
y1 = [ (y + (-1)**i*2*y ) for i in range(1,21,2) ]
y2 = [ (y + (-1)**(i+i)*(i*i)*y ) for i in range(1,21,2) ]
y3 = [ (y + (-1)**(i+i)*(-i*i)*y ) for i in range(1,21,2) ]

plt.plot(x,y2, label='upper')
plt.plot(x,y1, label='mid')
plt.plot(x,y3, label='lower')

plt.legend()
```
