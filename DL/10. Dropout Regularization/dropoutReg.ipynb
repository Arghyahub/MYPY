#### When using ANN, some neurons may be biased so , we can exclude those

```python
import numpy as np 
import keras
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df = pd.read_csv('./sonar_dataset.csv',header=None)
print("Classify Mines or Rocks using Sonar data")
df.sample(5)
```

```python
df.rename(columns={60:'Mine'},inplace=True)
df.sample(2)
```

```python
df['Mine'].replace({'R':0,'M':1},inplace=True)
df.sample(2)
```

```python
df.isna().sum()
```

```python
X = df.drop('Mine',axis='columns')
Y = df['Mine']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,Y,random_state=5,test_size=0.2)

X_train.shape, X_test.shape, y_train.shape,y_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=60,batch_size=8)
```

Too good to be true na, lets see how it performs

```python
model.evaluate(X_test,y_test)
```

Basically it's overfitted

```python
y_pred = model.predict(X_test)
y_pred = np.array([1 if y>=0.5 else 0 for y in y_pred])
y_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_pred,y_test))
```

### We can put drop out layer to drop some neurons to reduce overfitting

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=60,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```
