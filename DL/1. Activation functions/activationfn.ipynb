### Different activation functions

```python
from matplotlib import pyplot as plt
import math
import numpy as np

x = list(np.arange(-20,21,0.2))
def showAxes():
    # Add horizontal and vertical lines at x=0 and y=0
    plt.axhline(0, color='red', linewidth=0.5)  # Horizontal line at y=0
    plt.axvline(0, color='red', linewidth=0.5)  # Vertical line at x=0
```

##### Linear (y=x)

```python
showAxes() 
plt.plot(x,x)
plt.show()
```

##### Relu (y=max(0,x))

```python
def relu(n):
    return max(0,n)

showAxes() 
plt.plot(x,[relu(xi) for xi in x])
```

##### Sigmoid ( y = 1/(1+e^[-z]) ) where z=wx+b 

```python
def zVal(w,b,x):
    return w*x + b

def sigmoid(w,b,x):
    z = zVal(w,b,x)  
    return 1/(1+ math.exp(-z))

showAxes() 
plt.plot(x,[sigmoid(0.2,0.2,xi) for xi in x])
```

##### Tanh ( y = (e^z - e^[-z]) / (e^z + e^[-z]) )

```python
def zVal(w,b,x):
    return w*x + b

def E(n):
    return math.exp(n)

def tanh(w,b,x):
    z = zVal(w,b,x)  
    return (E(z) - E(-z)) / (E(z) + E(-z))

showAxes() 
plt.plot(x,[tanh(0.3,0.3,xi) for xi in x])
```
