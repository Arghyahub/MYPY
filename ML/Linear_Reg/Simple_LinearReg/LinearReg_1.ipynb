```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('./one_1.csv')
df
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price, color='red', marker='x')
```

```python
reg = linear_model.LinearRegression()

# First argument accepts a dataframe, [[]] return a new dataframe and [] return the series object

reg.fit(df[['area']],df.price)
```

```python
# Takes dataframe as input
pred = pd.DataFrame({ 'area': [3300] })
reg.predict(pred)
```

```python
w = reg.coef_   #slope
b = reg.intercept_  #intercept c

plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price, color='red', marker='x')

#plot graph
x = np.array(df.area) 
y = w*x + b
# plt.plot(x,y)
# or
plt.plot(x,reg.predict(df[['area']]))
```
