# Bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,5)
xname = ['GOOGL','AAPL','MSFT','FB']

y = [85,124,82,25]

profit = [34,80,33,14]

plt.xticks(x,xname)
plt.bar(x-0.2,y, width=0.4, color='blue', label='Revenue')
plt.bar(x+0.2,profit, width=0.4, color='green', label='Profit')
plt.legend()
```
